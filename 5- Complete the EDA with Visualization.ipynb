```python
# pandas is a software library written for the Python programming language for data manipulation and analysis.
import pandas as pd
#NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np
# Matplotlib is a plotting library for python and pyplot gives us a MatLab like plotting framework. We will use this in our plotter function to plot data.
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns
```

```python
## Exploratory Data Analysis

from js import fetch
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(dataset_part_2_csv)
df.head(5)
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
sns.scatterplot(data=df, x='FlightNumber', y='LaunchSite', hue='Class', palette='viridis')
plt.xlabel('Flight Number', fontsize=14)
plt.ylabel('Launch Site', fontsize=14)
plt.title('Relationship between Flight Number and Launch Site', fontsize=16)
plt.show()
```

```python
sns.catplot(x="FlightNumber", y="LaunchSite", hue="Class", data=df, aspect=3, height=6, kind="swarm")
plt.xlabel('Flight Number', fontsize=14)
plt.ylabel('Launch Site', fontsize=14)
plt.title('Flight Number vs Launch Site by Class', fontsize=16)
plt.show()
```

```python
sns.catplot(x="PayloadMass", y="LaunchSite", hue="Class", data=df, aspect=3, height=6, kind="swarm")
plt.xlabel('Payload Mass (kg)', fontsize=14)
plt.ylabel('Launch Site', fontsize=14)
plt.title('Payload Mass vs Launch Site by Class', fontsize=16)
plt.show()
```

```python
# Convert 'LaunchDate' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract year from 'LaunchDate'
df['Year'] = df['Date'].dt.year

# Calculate the proportion of successful launches for each year
yearly_success_rate = df.groupby('Year')['Class'].mean()

# Plotting the launch success yearly trend
plt.figure(figsize=(10, 6))
yearly_success_rate.plot(marker='o', color='b', linestyle='-')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Success Rate', fontsize=14)
plt.title('Launch Success Yearly Trend', fontsize=16)
plt.grid(True)
plt.show()
```

```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.year)
    return year

# Call the function to extract years
Extract_year()

# Create a new column 'Year' with the extracted years
df['Year'] = year

# Display the DataFrame
df.head()
```

```python
# Calculate the success rate for each year
yearly_success_rate = df.groupby('Year')['Class'].mean()

# Plotting the line chart
plt.figure(figsize=(10, 6))
yearly_success_rate.plot(marker='o', color='b', linestyle='-')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Success Rate', fontsize=14)
plt.title('Launch Success Yearly Trend', fontsize=16)
plt.grid(True)
plt.show()
```

```python
## Features Engineering

features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
# Create dummy variables for categorical columns
features = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

# Display the first few rows of the features DataFrame with dummy variables
features.head()
```

```python
# Cast all numeric columns to float64
features_one_hot = features.astype('float64')

# Display the first few rows of the features_one_hot DataFrame
features_one_hot.head()
```
